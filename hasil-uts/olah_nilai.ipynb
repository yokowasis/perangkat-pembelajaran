```python
# read csv
import pandas as pd

df = pd.read_csv('./rekap.csv') # type: ignore
# replace "-" with 0
df = df.replace('-', 0) # type: ignore
```

```python
prefix = [
    'Al-Quran_Hadis',
    'Aqidah_Akhlak',
    'Bahasa_Arab_Tingkat_Lanjut',
    'Bahasa_Arab',
    'Bahasa_Indonesia_Tingkat_Lanjut',
    'Bahasa_Indonesia',
    'Bahasa_Inggris_Tingkat_Lanjut',
    'Bahasa_Inggris',
    'Biologi',
    'Ekonomi_Pilihan',
    'Ekonomi',
    'Fikih',
    'Fisika',
    'Geografi',
    'Ilmu_Hadis',
    'Informatika',
    'Kimia',
    'Matematika_Tingkat_Lanjut',
    'Matematika',
    'Pendidikan_Pancasila',
    'Sejarah_Kebudayaan_Islam',
    'Sejarah',
    'Seni_Budaya',
    'Sosiologi',
    'Usul_Fikih'
]

# copy df
new_df = df.copy()
new_df
```

```python
rekap_df = pd.DataFrame()
# add column username, nama, kelas from df
rekap_df[['Username', 'Nama', 'Kelas']] = new_df[['Username', 'Nama', 'Kelas']]

for processed_prefix in prefix:
    grouped_cols = []
    grouped_df = pd.DataFrame()

    # loop through columns
    for col in new_df.columns:
        # save every cols that starts with prcessed_prefix
        if col.startswith(processed_prefix):
            grouped_cols.append(col)  # type: ignore

    # convert columns to numeric before calculating max
    grouped_df[processed_prefix] = new_df[grouped_cols].apply(
        pd.to_numeric, errors='coerce').max(axis=1)  # type: ignore

    # add group_df to rekap_df
    rekap_df = pd.concat([rekap_df, grouped_df], axis=1)

rekap_df
```

```python
# create a new df based on rekap_df with unique values in Kelas column
kelas_list = rekap_df['Kelas'].unique().tolist()
kelas_list.sort()

classes_df = []

for kelas in kelas_list:
    class_df = rekap_df[rekap_df['Kelas'] == kelas] # type: ignore
    classes_df.append(class_df) # type: ignore

# loop trough each df in classes_df and delete column that has all values 0
for i in range(len(classes_df)):
    classes_df[i] = classes_df[i].loc[:, (classes_df[i] != 0).any(axis=0)] # type: ignore
    classes_df[i] = classes_df[i].sort_values(by=['Nama']) # type: ignore
    classes_df[i] = classes_df[i].reset_index(drop=True) # type: ignore
    classes_df[i].index += 1 # type: ignore
    classes_df[i].to_csv(f'./rekap-{kelas}.csv', index_label='No') # type: ignore
```
